```python
import time
import matplotlib.pyplot as plt
import numpy as np
import cv2
from scipy.optimize import minimize

Filename = '1'

Image = np.float32(cv2.imread(Filename+'.tif',-1))
ImageSmooth = cv2.GaussianBlur(Image,(0,0),1.5,1.5)
#ImageHF = np.log(Image) - 
#ImageSmooth = cv2.bilateralFilter(Image, 9, 50, 50)

plt.rcParams["figure.figsize"] = (15,30)
plt.subplot(2,1,1)
plt.imshow(Image[2800:3000, 1000:1200])
plt.subplot(2,1,2)
plt.imshow(ImageSmooth[2800:3000, 1000:1200])
plt.show()
```

```python
ImageCorr = np.copy(Image)
ImageCorr[ImageCorr==0] = 1

StartRaw = int(ImageCorr.shape[0]/2)
IterN = 4

plt.rcParams["figure.figsize"] = (15,20)
plt.subplot(2,1,1)
plt.plot(ImageCorr[:,StartRaw])

t0 = time.time()
#DeltaMedians= np.zeros(ImageCorr.shape[1]-1)
# Идем из серидины вниз:
for r in range(StartRaw, ImageCorr.shape[0]-1):
#c = int(ImageCorr.shape[1]/2)+20
    # Итеративный подбор дельты для каждой пары соседних столбцов, приравн. медиану их отношения к 1:
    StepScale = 0.9
    Step = 0.
    ImageCorr[r,:] = ImageCorr[r,:]*np.median(ImageSmooth[r,:])/np.median(ImageCorr[r,:])
    #RawRatioMedianInit = np.median((ImageCorr[r+1,:] + Step)/ImageCorr[r,:])
    for i in range(IterN):
        RawRatioMedian = np.median((ImageCorr[r+1,:] + Step)/ImageCorr[r,:])
        Step += (1 - RawRatioMedian)*np.median(ImageCorr[r+1,:])*StepScale
        StepScale = StepScale*0.9
    #print('RawRatioMedianInit', RawRatioMedianInit, 
    #      'RawRatioMedian: ', np.median((ImageCorr[r+1,:] + Step)/ImageCorr[r,:]))
    ImageCorr[r+1,:] += Step
# Идем из серидины вверх:
for r in range(1, StartRaw)[::-1]:
    # Итеративный подбор дельты для каждой пары соседних столбцов, приравн. медиану их отношения к 1:
    StepScale = 0.9
    Step = 0.
    ImageCorr[r,:] = ImageCorr[r,:]*np.median(ImageSmooth[r,:])/np.median(ImageCorr[r,:])
    for i in range(IterN):
        RawRatioMedian = np.median((ImageCorr[r-1,:] + Step)/ImageCorr[r,:])
        Step += (1 - RawRatioMedian)*np.median(ImageCorr[r-1,:])*StepScale
        StepScale = StepScale*0.9
    ImageCorr[r-1,:] += Step
    
print(time.time() - t0)
plt.subplot(2,1,2)
plt.plot(ImageCorr[:,StartRaw])
#print(DeltaMedians)
cv2.imwrite(Filename+'_MedianFilt_Corr.tif', np.uint16(ImageCorr))
```

```python
start = time.time()
img_rec = np.copy(img_str)

# обрабатываем NaN
img_rec[np.isnan(img_rec)] = 0

# убираем соль и перец:
PepperVal = np.min(img_rec)
SoltVal = np.max(img_rec)
MeanVal = np.mean(img_rec)
for i in range(0,np.shape(img_rec)[0]):
    for j in range(0,np.shape(img_rec)[1]):
        if img_rec[i,j]==PepperVal or img_rec[i,j]==SoltVal:
            if i==0 or j==0 or i==np.shape(img_rec)[0] or j==np.shape(img_rec)[1]:
                img_rec[i,j] = MeanVal
            else:
                Area = img_rec[i-1:i+2,j-1:j+2]
                img_rec[i,j] = np.median(np.ravel(Area))
t1 = time.time()
print(t1 - start)
                
# удаляем полосы медианой:
img_rec_med = np.copy(img_rec)
for i in range(img_rec_med.shape[1]-1):
    fun = lambda x: np.abs(1 - np.median(img_rec_med[0::10,i]/(x[0]*img_rec_med[0::10,i+1] + x[1])))
    bnds = ((-100, 100), (-100, 100))
    res = minimize(fun, (1, 0), bounds=bnds, tol=1e-2)
    img_rec_med[:,i+1] = res.x[0]*img_rec_med[:,i+1] + res.x[1]
t2 = time.time()
print(t2 - t1)

# удаляем полосы итерационным усреднением:
img_rec_mean = np.copy(img_rec)
for i in range(img_rec_mean.shape[1]-1):
    Col_i = img_rec_mean[0::10,i]
    Col_ip1 = img_rec_mean[0::10,i+1]
    ColRatio = Col_i/Col_ip1
    ColRatio_Mean = np.mean(ColRatio)
    ColRatio_Std = np.std(ColRatio)
    Col_i = Col_i[ColRatio>=ColRatio_Mean-2*ColRatio_Std]
    Col_i = Col_i[ColRatio<=ColRatio_Mean+2*ColRatio_Std]
    Col_ip1 = Col_i[ColRatio>=ColRatio_Mean-2*ColRatio_Std]
    Col_ip1 = Col_i[ColRatio<=ColRatio_Mean+2*ColRatio_Std]
    fun = lambda x: np.abs(1 - np.mean(img_rec_mean[0::10,i]/(x[0]*img_rec_mean[0::10,i+1] + x[1])))
    bnds = ((-100, 100), (-100, 100))
    res = minimize(fun, (1, 0), bounds=bnds, tol=1e-2)
    img_rec_mean[:,i+1] = res.x[0]*img_rec_mean[:,i+1] + res.x[1]
t3 = time.time()
print(t3 - t2)
```
